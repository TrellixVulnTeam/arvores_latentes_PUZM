```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
parsed = []
with open("predict_arcs.log", "r") as f:
    for line in f:
        if "Embeddings" not in line:
            continue
        blocks = line.split(":")
        values = blocks[1].replace("[","").replace("]","").split()
        flag = blocks[2]
        parsed.append([float(values[0]), float(values[1]), "True" in flag])
```

```python
df = pd.DataFrame(parsed, columns=[0,1,"Label"])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

colors = ['r' if x else 'b' for x in df["Label"]]
sizes = [3 if x else 1 for x in df["Label"]]
ax.scatter(np.clip(df[0],0,6), np.clip(df[1],0,6), s=sizes, c=colors)

lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]

# now plot both limits against eachother
ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
ax.set_aspect('equal')
ax.set_xlim(lims)
ax.set_ylim(lims)

plt.show()
```

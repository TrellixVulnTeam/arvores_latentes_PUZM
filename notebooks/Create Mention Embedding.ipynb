```
import os
import re
from tqdm import tqdm
os.environ["BERT_DB_PATH"]="data_in/sliding"
from data_handling import create_mention_embedding_ts
```

```
def get_file_name(out_folder, doc_id):
    name_re = re.compile(r".*\((.*)\); part (\d\d\d)")
    match = name_re.match(doc_id)
    file_name = f"{match.group(1)}_{match.group(2)}".replace("/", "_")
    subfolders = file_name.split("_")[:3]
    final_name = os.path.join(out_folder, *subfolders, file_name + ".tsv")
    o_name = os.path.join(out_folder, file_name + ".tsv")
    #print(f"##### {final_name}")
    return final_name, o_name

def check_all_exist(file_name, check_dir):
    with open(file_name, "r") as f:
        current_file = []
        to_check = ""
        for line in f:
            if "#begin" in line:
              to_check, to_check2 = get_file_name(check_dir, line.replace("#begin ", ""))
              current_file.clear()
            elif "#end" in line:
              if (not os.path.isfile(to_check)) and \
                  (has_annotations(current_file)) and \
                  (not os.path.isfile(to_check2)):
                return False
            else:
                current_file.append(line)
    return True
  
def has_annotations(lines):
  for line in lines:
    #print(f"Line:{line} [{len(line)}]")
    if (len(line) > 10) and (line.split()[-1] != "-"):      
      return True
  return False
```

```
out_folder = r"data_out"
dst_folder = "/content/gdrive/My Drive/ProjetoFinal/tsv/cumsum_40/"

for root, dirs, files in tqdm(list(os.walk("data_in/train/"))):
  for f in files:
    in_file = f"{root}/{f}"
    print(in_file)
    if check_all_exist(in_file, dst_folder):
      #print("All generated. Ignoring")
      continue    
    create_mention_embedding_ts.create_ts_file(in_file, out_folder, use_annotated=True, as_diag=True)
    for file_name in os.listdir(out_folder):
      levels = file_name.split("_")[:3]
      dest = os.path.join(dst_folder,*levels)
      print(f"Moving {file_name} to {dest}")
      !mkdir -p '{dest}'
      !mv '{out_folder}/{file_name}' '{dest}/'
```

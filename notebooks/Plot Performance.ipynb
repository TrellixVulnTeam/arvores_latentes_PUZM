```python
import pandas as pd
import pickle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open("stats.dmp", "rb") as f:
    df = pickle.load(f)

df["composite_index"] = df["epoch"] * 1000 + df["step"]
pivot = df.pivot("composite_index", "name", "conll")
```

```python
ax = pivot.reset_index(drop=True).rolling(5).mean().plot(figsize=(15,5),style=['-','s-','o-','^-'],)
_ = plt.title("Performance no dataset de mini-validação")
_ =plt.xlabel("Etapa")
_ = plt.ylabel("Score CoNLL - Média móvel de 10 etapas")
```

```python
ax = pivot.reset_index(drop=True)["baseline"].plot(figsize=(15,5))
_ = plt.title("Performance no dataset de mini-validação - Baseline")
_ =plt.xlabel("Etapa")
_ = plt.ylabel("Score CoNLL")
```

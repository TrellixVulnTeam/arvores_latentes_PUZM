```
with tf.Session() as sess:
  new_saver = tf.train.import_meta_graph(f'{checkpoint}.meta')
  save_path = tf.train.latest_checkpoint(f"{root}/models/cased_L-12_H-768_A-12/",latest_filename="bert_model.ckpt.data-00000-of-00001")
  new_saver.restore(sess, save_path)
```

```
# import the inspect_checkpoint library
from tensorflow.python.tools import inspect_checkpoint as chkp

# print all tensors in checkpoint file
chkp.print_tensors_in_checkpoint_file(checkpoint, tensor_name='', all_tensors=False, all_tensor_names=True)
```
